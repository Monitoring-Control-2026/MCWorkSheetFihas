## Виконання роботи

### Підготовка середовища

Для початку потрібно імпортувати потрібні бібліотеки, які необхідні для роботи. Numpy - для обчислень, а pandas - для роботи з данними.

```python
import numpy as np
import pandas as pd
```

Далі використаємо таку саму матрицю виграшів, що і в попередній лабораторній роботі.

```python
student_alternatives = [
    "Моніторинг ситуації",
    "Обмеження доступу",
    "Резервна інфраструктура",
    "Блокування інфраструктури"
]

student_states = [
    "Відсутність загрози",
    "Низька загроза",
    "Середня загроза",
    "Велика загроза"
]

student_matrix = np.array([
    [10,  4, -2, -10],
    [ 6,  7,  1, -5],
    [ 1,  3,  7,  2],
    [ -7, -4 , 3, 10]
], dtype=float)

student_probabilities = np.array([0.40, 0.30, 0.25, 0.05], dtype=float)

student_df = pd.DataFrame(student_matrix, index=student_alternatives, columns=student_states)
student_df
```

### Критерій Гурвіца

Далі перейдемо до перевірки критеріїв. Почнімо перевірку з критерію Гурвіца. Даний критерій використовує показник песимізму $alpfa$, де чим більше значення, ти більш песимістичним розрахунок буде. Для роботи використаємо проміжок від 0 до 1 з кроком 0.2. З результатів можна помітити що чим більше даний показник, тим сильніше альтернатива хилиться у бік резервної інфраструктури.

```python
def hurwicz_criterion(matrix, alternatives, alpha=0.5):
    row_mins = matrix.min(axis=1)
    row_maxs = matrix.max(axis=1)
    scores = alpha * row_mins + (1 - alpha) * row_maxs
    best_index = scores.argmax()

    result_df = pd.DataFrame({
        "Альтернатива": alternatives,
        "min по рядку": row_mins,
        "max по рядку": row_maxs,
        "Оцінка Гурвіца": scores
    })
    return result_df, alternatives[best_index], scores[best_index]
    
for alpha in np.arange(0, 1.1, 0.2):
    hurwicz_df, hurwicz_best_alt, hurwicz_best_value = hurwicz_criterion(
        student_matrix,
        student_alternatives,
        alpha=alpha
    )

    display(hurwicz_df)
    print(f"Параметр alpha = {alpha}")
    print(f"Оптимальна альтернатива за критерієм Гурвіца: {hurwicz_best_alt}")
    print(f"Оцінка критерію: {hurwicz_best_value:.3f}")
```

### Критерій Ходжа-Лемана

Наступним критерієм для перевірки буде критерій Ходжа–Лемана, який дозволяє поєднати довіру до ймовірностей і обережність щодо найгіршого сценарію. Для цього потребується додаткова величина $v$, яка означає довіра розподілу. Для завдання також візьмемо значення від 0 до 1 з кроком 0.2. У результаті перевірки критерію вияснилось, що поки довіра не є максимальною, використовується резервна інфраструктура, а при максимальній довіри використовується обмеження доступу.

```python
def hodge_lehmann_criterion(matrix, alternatives, probs, v=0.5):
    row_mins = matrix.min(axis=1)
    expected_values = matrix @ probs
    scores = v * expected_values + (1 - v) * row_mins
    best_index = scores.argmax()

    result_df = pd.DataFrame({
        "Альтернатива": alternatives,
        "min по рядку": row_mins,
        "Математичне сподівання": expected_values,
        "Оцінка Ходжа–Лемана": scores
    })
    return result_df, alternatives[best_index], scores[best_index]
    
for v in np.arange(0, 1.1, 0.2):
    hl_df, hl_best_alt, hl_best_value = hodge_lehmann_criterion(
        student_matrix,
        student_alternatives,
        student_probabilities,
        v=v
    )

    display(hl_df)
    print(f"Параметр v = {v}")
    print(f"Оптимальна альтернатива за критерієм Ходжа–Лемана: {hl_best_alt}")
    print(f"Оцінка критерію: {hl_best_value:.3f}")
```

### Критерій Гермейєра

Наступним критерієм для аналізу буде критерій Гермейєра, який застосовується у матриці збитків, тобто матриці мають бути повністю від'ємні. Оптимальним варіантом виявився обмеження доступу, оскільки при його використанні збитків буде найменше.

```python
def hermeyer_criterion(matrix, alternatives, probs):
    if np.any(matrix > 0):
        shift = matrix.max() + 1
        transformed_matrix = matrix - shift
    else:
        transformed_matrix = matrix.copy()

    weighted = transformed_matrix * probs
    row_mins = weighted.min(axis=1)
    best_index = row_mins.argmax()

    transformed_df = pd.DataFrame(transformed_matrix, index=alternatives, columns=student_states)
    weighted_df = pd.DataFrame(weighted, index=alternatives, columns=student_states)
    summary_df = pd.DataFrame({
        "Альтернатива": alternatives,
        "min(p_j * e_ij)": row_mins
    })
    return transformed_df, weighted_df, summary_df, alternatives[best_index], row_mins[best_index]

hermeyer_transformed_df, hermeyer_weighted_df, hermeyer_summary_df, hermeyer_best_alt, hermeyer_best_value = hermeyer_criterion(
    student_matrix,
    student_alternatives,
    student_probabilities
)

print("Перетворена матриця для критерію Гермейєра:")
display(hermeyer_transformed_df)

print("Зважені значення p_j * e_ij:")
display(hermeyer_weighted_df)

print("Підсумкова таблиця:")
display(hermeyer_summary_df)

print(f"Оптимальна альтернатива за критерієм Гермейєра: {hermeyer_best_alt}")
print(f"Оцінка критерію: {hermeyer_best_value:.3f}")
```

### Критерій добутку

Достанній критерій який буде використаний це критерій добутку. Для його використання спочатку треба змінити таблицю у додатній вигляд. Провевши розрахунки, виявилось що резервна інфраструктура буде найоптимальнішим варіантом.

```python
def product_criterion(matrix, alternatives):
    # Для критерію добутків потрібні додатні значення.
    if np.any(matrix <= 0):
        shift = abs(matrix.min()) + 1
        transformed_matrix = matrix + shift
    else:
        transformed_matrix = matrix.copy()

    row_products = transformed_matrix.prod(axis=1)
    best_index = row_products.argmax()

    transformed_df = pd.DataFrame(transformed_matrix, index=alternatives, columns=student_states)
    summary_df = pd.DataFrame({
        "Альтернатива": alternatives,
        "Добуток елементів рядка": row_products
    })
    return transformed_df, summary_df, alternatives[best_index], row_products[best_index]

product_transformed_df, product_summary_df, product_best_alt, product_best_value = product_criterion(
    student_matrix,
    student_alternatives
)

print("Перетворена матриця для критерію добутків:")
display(product_transformed_df)

print("Підсумкова таблиця:")
display(product_summary_df)

print(f"Оптимальна альтернатива за критерієм добутків: {product_best_alt}")
print(f"Оцінка критерію: {product_best_value:.3f}")
```

### Математичне сподівання і ризик результату

На останок буде перевірено значення очікуваного виграшу, дисперсію та середньоквадратичне відхилення в усіх альтернативах.
Серед результатів можна помітити що найстабільнішим є варіант резервної інфраструктури, оскільки дисперсія найменша.

```python
def expected_value_and_risk(matrix, alternatives, probs):
    expected_values = matrix @ probs
    variances = ((matrix - expected_values.reshape(-1, 1)) ** 2 @ probs)
    stds = np.sqrt(variances)

    result_df = pd.DataFrame({
        "Альтернатива": alternatives,
        "Математичне сподівання": expected_values,
        "Дисперсія": variances,
        "Середньоквадратичне відхилення": stds
    })
    return result_df
risk_df = expected_value_and_risk(student_matrix, student_alternatives, student_probabilities)
risk_df
```

### Підсумкове порівняння критеріїв

Для зручності всі результати зведемо до однієї таблиці.

```python
summary_df = pd.DataFrame({
    "Критерій": ["Гурвіца", "Ходжа–Лемана", "Гермейєра", "Добутків"],
    "Оптимальна альтернатива": [
        hurwicz_best_alt,
        hl_best_alt,
        hermeyer_best_alt,
        product_best_alt
    ],
    "Оцінка": [
        round(hurwicz_best_value, 3),
        round(hl_best_value, 3),
        round(hermeyer_best_value, 3),
        round(product_best_value, 3)
    ]
})

summary_df
```

# Висновки

У цій лабораторній роботі було проведено аналіз матриці виграшів на тему "Стратегія реагування на інцедент" розміром 4 на 4. Для кожного зі станів у даній матриці було додано свою ймовірність, де на кращу подію ймовірність найвища і спадала сильніще в залежності від критичності ситуації.
Порівнюючи отримані в цій лабораторній роботі результати з результатами попередньої роботи, можна знайти певну параллель. В критерії Гурвіца можна помітити результати як критерію оптимізму, як і критерію Вальда. Тобто показник $\alpha$ вказує на те, наскільки песимістично чи оптимістично особа дивиться на ситуацію, що вприлає на стратегію вибору. Також не варто уникати критерій Ходжа–Лемана, який також має результати з критерія Вальда та Байєса-Лапласа. Показник $v$ вказує на рівень довіри до ймовірностей. В критерії Гермейєра найоптимальнішим варіантом виявилось обмеження обмеження оступу, а критерії добутку - резервна інфраструктура.
Було спостережено що серед усіх критеріїв, в даній ситуації, варіант з резервною інфраструктурою зустрічався найчастіше, після нього обмеження доступу.
Що до використання переглянутих критеріїв, то критерій Гурвіца дозволяє переглянути ситуацію не тільки як оптиміст або песиміст, але і десь посередині; критерій Ходжа-Лемана дозволяє переглянути наскільки збільшиться чи зменшиться прибуток якщо додати до роботи математичне сподівання; критерій Гермейера дозволяє перевірити наскільки високою буде втрата при виборі певного варіанту, що дасть певні знання що до втрат.

# Контрольні питання

 Який зміст має параметр $\alpha$ у критерії *Гурвіца*?

 Параметр $\alpha$ вказує на рівень песемизму, при якому чим ближче до одиниці, тим сильніше критерій Гурвіца схиляється до найменших втрат, тобто до логіки Вальда, в противному випадку -- до логіки максімакс.

У чому полягає зміст параметра $v$ у критерії *Ходжа–Лемана*?

Параметр $v$ вказує на ступінь довіри до розподілу ймовірностей, де чим ближче параметр до одиниці, тим більше Критерій схожий на на критерій Байєса-Лапласа, в інакшому випадку до критерію Вальда.

За яких умов застосовують критерій *Гермейєра*?



Чому для критерію добутків потрібні *додатні* значення?

Тому що при множені на негативне число чи нуль результат може виявитись недостовірним. Наприклад при множені двох поганих значень вийде неймовірно позитивний результат, що є малоймовірним. Або якщо одним зі значеннь буде нуль то попри старання результат вийде ніяким, що теж малоймовірно.

Чим відрізняється аналіз *очікуваного виграшу* від аналізу *ризику*?

Аналіз очікуваного виграшу орієнтується на кількість прибутку чи виграшу в певній ситуації, тоді як аналіз ризику орієнтується на кількість втрат в певній ситуації. Враховуючи це то і вибір критерію для того чи іншого аналізу буде вирізнятись

Чому високе математичне сподівання не завжди означає найкраще рішення?

Тому що в математичному сподіванні є ризик того що можна багато втратити навіть з мінімальною ймовірністю. Воно показує середній виграш чи програш при багаторазовому використанні математичного сподівання.

Як змінюється вибір альтернативи при зростанні песимізму ОПР?

При зростанні песимізму ОПР вибір альтернативи змінюється у бік зменшення втрат, ніж до збільшення прибутку.
